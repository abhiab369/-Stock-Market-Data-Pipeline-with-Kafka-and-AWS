```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import pandas as pd
```

```python
consumer = KafkaConsumer('demo',
                        bootstrap_servers=['3.6.41.228:9092'],
                        value_deserializer = lambda x:loads(x.decode('utf-8')))
```

```python
from s3fs import S3FileSystem
```

```python
#before using s3 bucket in jupyter we must have to setup aws in machine using cli
s3 = S3FileSystem()
```

```python
for count,i in enumerate(consumer):
    with s3.open('s3://abhishek-kafka/_{}.json'.format(count),'w') as file:
        json.dump(i.value, file)
        
```

```python
for count,i in enumerate(consumer):
    print(count)
    print(i.value)
```

```python
import boto3
import json

s3 = boto3.client('s3')

for count,i in enumerate(consumer):
    object_key = 'stock_market_{}.json'.format(count)
    data = json.dumps(i.value)
    s3.put_object(Body=data, Bucket = 'abhishek-kafka',Key=object_key)
```
